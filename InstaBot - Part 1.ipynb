# Function for login

```python
def login():
    
    #go to the website
    driver.get("https://www.instagram.com")
    time.sleep(3)
    
    #find the username box and enter your username
    userName = driver.find_element_by_name('username')
    userName.send_keys("SAMPLE USERNAME")
    
    #find the password box and enter your password
    password = driver.find_element_by_name('password')
    password.send_keys('SAMPLE PASSWORD')
    
    #find the submit button and submit
    btn = driver.find_element_by_class_name('sqdOP')
    btn.submit()
    
login()
```

# Typing 'food' in the search bar and printing all the account names displayed

```python
def searchFood():

    #go to the website
    driver.get("https://www.instagram.com")
    
    #find the search bar and type in 'food'
    search = driver.find_element_by_class_name('XTCLo')
    search.send_keys('food')
    time.sleep(3)

    #get the list of all the account displayed
    li = driver.find_elements_by_class_name("Ap253")
    for acc in li:
        #check if the account is not a hashtag
        if acc.text[0] != '#':
            print(acc.text)

searchFood()
```

# Create functions for following and unfollowing

```python
def follow():
    
    #find the follow button
    button = driver.find_element_by_xpath('//span[@class="vBF20 _1OSdk"]/button')
    
    #if the account is not already followed, then press the follow button
    if button.text == 'Follow':
        button.click()
        print('Followed')
        return
    
    #if the account is already followed
    print('Already Followed')
    
def unfollow():
    
    #find the follow button
    button = driver.find_element_by_xpath('//span[@class="vBF20 _1OSdk"]/button')
    
    #if the account is already unfollowed
    if button.text == 'Follow':
        print('Already unFollowed')
        return
    
    #if the account is not already unfollowed, then click on the button
    button.click()
    time.sleep(2)
    
    #select the unfollow popup and click the unfollow button
    unfollow = driver.find_element_by_class_name('-Cab_')
    unfollow.click()
    print('Unfollowed')
    

def soDelhi():
    
    #open an example account --> SoDelhi
    openSoDelhi()
    time.sleep(3)
    
    #follow the account
    follow()
    time.sleep(2)
    
    #unfollow the account
    unfollow()
    time.sleep(2)
    

soDelhi()   
    
    
```

# A function to like and unlike the posts

```python
def likeUnlike(toBeLiked):

    #go to the website
    driver.get("https://www.instagram.com")
    time.sleep(3)
    
    #open an example account -> dilsefoodie
    search = driver.find_element_by_class_name('XTCLo')
    search.send_keys('dilsefoodie')
    time.sleep(3)
    account = driver.find_element_by_xpath('//div[@class="fuqBx"]/a[1]')
    account.click()
    time.sleep(3)
    
    #scroll the window
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(2)

    #get the lists of posts
    posts = driver.find_elements_by_class_name("eLAPa")

    #iterate over the first 30 posts
    for i in range(30):

        #click on the post
        post = posts[i]
        post.click()
        time.sleep(2)

        #find the like button
        like = driver.find_element_by_xpath("//span[@class = 'fr66n']/button")
        svg = driver.find_element_by_css_selector(".fr66n .QBdPU svg")
        
        #conditions for clicking the like button
        if (toBeLiked==True and svg.get_attribute('aria-label') == 'Like') or (toBeLiked==False and svg.get_attribute('aria-label')=='Unlike'):
            like.click()
            
        else:
            
            #if the post is already liked
            if toBeLiked == True:
                print('post', i+1, 'is already liked')
                
            #if the post is already unliked
            else:
                print('post', i+1, 'is already unliked')
            
        

        close = driver.find_element_by_xpath("//body/div[last()]/div[last()]/button")
        close.click()
        time.sleep(2)
    

#pass True to like the posts and False to unlike the posts
likeUnlike(toBeLiked = False)
    
    
```

# Functions to print Mutual Followers

```python
def getMutualFollowers():
    
    driver.find_element_by_css_selector("span.tc8A9").click()
    time.sleep(2)
    driver.find_element_by_link_text("See All Followers").click()
    time.sleep(2)

    fBody = driver.find_element_by_class_name("isgrP")
    arr = driver.find_elements_by_css_selector(".isgrP li")
    i = 0

    mutual = set()

    while True:

        try:
            account = arr[i]

            if account.text.split("\n")[-1] == "Following":
                mutual.add(account.text.split("\n")[0])
                i += 1
            else:
                break

        except IndexError:

            driver.execute_script("arguments[0].scrollBy(0, 1000);", fBody)
            time.sleep(2)
            arr = driver.find_elements_by_css_selector(".isgrP li")
            
    return mutual

def getMyFollowers():
    
    driver.get("https://www.instagram.com")
    time.sleep(2)
    driver.find_element_by_css_selector(".Fifk5 img._6q-tv").click()
    time.sleep(2)
    driver.find_element_by_css_selector("._01UL2 a").click()
    time.sleep(2)
    
    numberOfFollowers = int(driver.find_element_by_css_selector("a._81NM2 span").text)
    followers = driver.find_element_by_css_selector(".LH36I a").click()
    time.sleep(2)

    fBody = driver.find_element_by_class_name("isgrP")
    arr = driver.find_elements_by_css_selector(".isgrP li")


    myFollowers = set()
    i = 0

    while i < numberOfFollowers:

        try:
            account = arr[i]   
            myFollowers.add(account.text.split("\n")[0])
            i += 1


        except IndexError:

            driver.execute_script("arguments[0].scrollBy(0, 10000);", fBody)
            time.sleep(2)
            arr = driver.find_elements_by_css_selector(".isgrP li")

    return myFollowers

def getCommon():

    driver.get("https://www.instagram.com")

    search = driver.find_element_by_class_name('XTCLo')
    search.send_keys('foodtalkindia')
    time.sleep(3)

    account = driver.find_element_by_xpath('//div[@class="fuqBx"]/a[1]')
    account.click()
    time.sleep(3)
    
    mutual = getMutualFollowers()
    time.sleep(1)
    myFollowers = getMyFollowers()
    
    print("The required followers are: ")
    
    for name in mutual.difference(myFollowers):
        print(name)
        
        
getCommon()
        

```

# Function to see story

```python
def seeStory():

    driver.get("https://www.instagram.com")
    
    search = driver.find_element_by_class_name('XTCLo')
    search.send_keys('coding.ninjas')
    time.sleep(3)

    account = driver.find_element_by_xpath('//div[@class="fuqBx"]/a[1]')
    account.click()
    time.sleep(3)
    
    
    img = driver.find_element_by_class_name("RR-M- ")
    
    #if the user has no story
    if img.get_attribute('aria-disabled') == 'true':
        print('no story')
        return

    else:
        canvas = driver.find_element_by_css_selector(".RR-M- .CfWVH")
        
        #if story is already seen
        if canvas.get_attribute('width') == '87':
            print('already seen')
            return
        
        #if story is not seen
        else:
            img.click()
            time.sleep(3)
            print('viewed the story')
            return
        
seeStory()
    
```
